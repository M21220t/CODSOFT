```python
# My Task3(To-Do List)
tasks = []
def display_list():
    print(" "+"To-Do List:")
    for i, task in enumerate(tasks, 1):
        print(f"{i}. {task}")
    print()
def add_task_to_list(task):
    tasks.append(task)
    print(f"Task '{task}' added to the list.\n")
def remove_task_from_list(task_index):
    if task_index >= 1 and task_index <= len(tasks):
        removed_task = tasks.pop(task_index - 1)
        print(f"Task '{removed_task}' removed from the list.\n")
    else:
        print("Invalid task index. Please try again.\n")
while True:
    print("Options:")
    print("1. Display To-Do List stored.")
    print("2. Add your task to the list.")
    print("3. Remove a Task from  the list.")
    print("4. End  program")
    
    choice = input("Enter your choice: ")

    if choice == "1":
        display_list()
    elif choice == "2":
        new_task = input("Enter the task: ")
        add_task_to_list(new_task)
    elif choice == "3":
        display_list()
        task_index = int(input("Enter the task number to remove: "))
        remove_task_from_list(task_index)
    elif choice == "4":
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.\n")
```
